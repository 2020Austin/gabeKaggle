Create a Bar Plot of the number of TRUE(1) and FALSE(0) labels in the dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Load Data

```python
df = pd.read_csv("../cs-training.csv")
df.drop("Unnamed: 0", axis=1, inplace=True)
df.dropna(inplace=True)
```

Create counter object

```python
from collections import Counter
```

```python
c = Counter(df["SeriousDlqin2yrs"])
c = {"Yes":c[1], "No":c[0]}

for label, count in c.items():
       print(f"{label} -> {count}")
```

Draw the Barplot

```python
plt.figure(dpi=600) # sets resolution of plot
ax = sns.barplot(
       x=list(c.keys()), 
       y=list(c.values()),
       capsize = 0.05,
       palette="magma",
       
)

ax.set(xlabel='Serious Finances in 2 years', ylabel='Count', title="Proportion of each label in Training Dataset")

plt.show()
```
